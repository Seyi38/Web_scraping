```python
#import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requestsfrom selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import visibility_of_element_located
```

```python
#!pip install webdriver-manager
```

```python
#Start Chrome web browser
browser = webdriver.Chrome() 
#Get URL
browser.get("https://www.researchgate.net/profile/Consolee-Mbarushimana/publications")  # navigate to URL
#Wait for the page to load completely
```

```python
#Wait for <h2> element to appear on the page
title = (
    WebDriverWait(driver=browser, timeout=100)
    .until(visibility_of_element_located((By.CSS_SELECTOR, "h2")))
    .text
)
# retrieve fully rendered HTML content and save to content variable
content = browser.page_source
browser.close()
```

```python
#Parse html content using BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')
#Find all div classes specified below
publications = soup.find_all("div", class_="nova-legacy-v-publication-item--size-m")
#Create empty list to store data
data = []
for pub in publications:
    title_element = pub.find("div", class_="nova-legacy-v-publication-item__title")
    if title_element:
        title = title_element.find("a").text.strip()
    else:
        title = "N/A"
    
    date_element = pub.find("span", class_="")
    if date_element:
        date = date_element.text.strip()
    else:
        date = "N/A"
    
    author_elements = pub.find_all("span", class_="nova-legacy-v-person-inline-item__fullname")
    authors = [author.text.strip() for author in author_elements]
    
    description_element = pub.find("div", class_="nova-legacy-v-publication-item__description")
    if description_element:
        description = description_element.text.strip()
    else:
        description = "N/A"
    
    print("Title:", title)
    print("Date:", date)
    print("Authors:", ", ".join(authors))
    print("Description:", description)
    print()
    #add to data list
    data.append({"Title": title, "Date": date, "Authors": ", ".join(authors), "Description": description})
#convert to Pandas DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# Convert the DataFrame to a JSON file
df.to_json("publications.json", orient="records", indent=4)
print("JSON file saved.")
```
